```python
import numpy as np
import pandas as pd


df = pd.read_csv('data/bladedat.txt', sep='\t', names=['r', 'beta', 'c', 'thick'])
df
```

```python
from bem import steady_bem

R = 89.17
V_0 = 10.0
omega = 8.032*2*np.pi/60  # 8.032 RPM to rad/s
theta_p = 0
lamda = omega*R/V_0

# R = 89.17
# V_0 = 25.0
# omega = 9.6*2*np.pi/60  # 8.032 RPM to rad/s
# theta_p = 22.975
# lamda = omega*R/V_0

P_out, Thrust, C_P, C_T, C_P_local_list, C_T_local_list = steady_bem(R, V_0, omega, theta_p, lamda, df)


print(f'Thrust = {Thrust:.1f}')
print(f'C_T = {C_T:.3f}')
print(f'P_out = {P_out:.1f}')
print(f'C_P = {C_P:.3f}')
```

```python
import matplotlib.pyplot as plt

plt.plot(df['r'], C_P_local_list, '--', label='Original')
plt.xlabel('Radius [m]')
plt.ylabel('Local Power Coefficient')
plt.xlim(0, 90)
plt.ylim(-0.2, 0.8)
plt.grid(linestyle='--', linewidth=0.7, alpha=0.8)
plt.legend()
```
